## How to Plot Multiple PPCs on a Single Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#First Scenario
max_A1 = 100
max_B1 = 80
A1 = np.linspace(0, max_A1, 100)
B1 = (max_B1 * (max_A1 - A1)) / max_A1

#Second Scenario
max_A2 = 90
max_B2 = 70
A2 = np.linspace(0, max_A2, 100)
B2 = (max_B2 * (max_A2 - A2)) / max_A2

#Third scenario
max_A3 = 110
max_B3 = 90
A3 = np.linspace(0, max_A2, 100)
B3 = (max_B3 * (max_A3 - A3)) / max_A3

#Plotting
plt.figure(figsize=(10, 6))

#plot first PPC
plt.plot(A1, B1, color = "red", label = "PPC 1",)
plt.fill_between(A1, B1, color = "red", alpha = 0.3)

#plot second ppc
plt.plot(A2, B2, label = "PPC 2", color = "green")
plt.fill_between(A2, B2, color = "green", alpha = 0.3)

#plot third ppc
plt.plot(A3, B3, label = "PPC 3", color = "blue")
plt.fill_between(A3, B3, color = "blue", alpha = 0.3)

#fixing 

plt.title("Production Possibility Curves")
plt.xlabel("Good A Production")
plt.ylabel("Good B Production")
plt.xlim(0, max(max_B1, max_B2, max_B3))
plt.ylim(0, max(max_A1, max_A2, max_A3))
plt.grid(True)
plt.legend()
plt.show()
```

### Understanding Function Definitions

When defining a function, you can specify parameters (also known as arguments) in parentheses. If there are no parameters, the function simply executes the code provided. The colon (`:`) at the end of the function definition is mandatory, and the indentation that follows (typically four spaces or a tab) indicates the block of code that belongs to that function.

#### Example with Parameters

Here's another example where we include a parameter:

```python
def greet(name):
    print("Hello", name)
    
greet("Alice")
```

# Using functions to produce three production possibility curves

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
max_A = 100
max_B = 80

#function
def calculate_ppc(max_A, max_B):
    A = np.linspace(0, max_A, 100)
    B = (max_B * (max_A - A)) / max_A
    return A, B

# three scenarios
scenarios = [(100, 80), (90, 70), (110, 90)]
colors = ['red', 'green', 'blue']

plt.figure(figsize=(10, 6))

for i, (max_A, max_B) in enumerate(scenarios):
    A, B = calculate_ppc(max_A, max_B)
    plt.plot(A, B, label = f"PPC {i + 1}", color = colors[i])
    plt.fill_between(A, B, color = colors[i] , alpha = 0.3)
    
plt.title("Production Possibility Curves")
plt.xlabel("Good A Production")
plt.ylabel("Good B Production")
plt.xlim(0, max(max_A, max_A in scenarios))
plt.ylim(0, max(max_B, max_B in scenarios))
plt.grid(True)
plt.legend()
plt.show()
```
