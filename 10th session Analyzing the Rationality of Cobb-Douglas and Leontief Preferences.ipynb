# Analyzing the Rationality of Cobb-Douglas and Leontief Preferences

- We are going to analyze whether Cobb-Douglas and Leontief preferences satisfy the axioms of rationality or not.

# Cobb-douglas Preferences

$$ U(x, y) = x^{\alpha} y^{1-\alpha}$$

```python
def Cobb_Douglas(x, y):
    alpha = 0.5
    return x ** alpha * y ** 1 - alpha
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1 ,1000)  
y = np.arange(1 ,1000)  

X, Y = np.meshgrid(x, y)
Z = Cobb_Douglas(X, Y)

fig = plt.figure(figsize = (12, 5))

ax1 = fig.add_subplot(121, projection = "3d")
ax1.plot_surface(X, Y, Z, cmap = "Wistia")
ax1.set_title("Cobb_Douglas Utility Function")
ax1.set_xlabel("Good X")
ax1.set_ylabel("Good Y")
ax1.set_zlabel("Utility Levels")
ax1.grid(False)

ax2 = fig.add_subplot(122)
contour = plt.contour(X, Y, Z, levels=5, cmap='spring')
plt.colorbar(contour)
ax2.set_title("Contour Plot of Cobb_Douglas Utility Function")
ax2.set_xlabel("Good X")
ax2.set_ylabel("Good Y")
ax2.grid(False)

plt.tight_layout()
plt.show()
```

# Leontief Preferences

$$ U(x, y)= min(\frac{x}{w_1}, \frac{y}{w_2})$$

```python
def Leontief(x, y):
    w1 = 1
    w2 = 1
    return np.minimum(w1 * x, w2 * y)
```

```python
x = np.arange(1 ,1000)  
y = np.arange(1 ,1000)  

X, Y = np.meshgrid(x, y)
Z = Leontief(X, Y)

fig = plt.figure(figsize = (12, 5))

ax1 = fig.add_subplot(121, projection = "3d")
ax1.plot_surface(X, Y, Z, cmap = "viridis")
ax1.set_title("Leontief Utility Function")
ax1.set_xlabel("Good X")
ax1.set_ylabel("Good Y")
ax1.set_zlabel("Utility Levels")
ax1.grid(False)

ax2 = fig.add_subplot(122)
contour = plt.contour(X, Y, Z, levels=5, cmap='spring')
plt.colorbar(contour)
ax2.set_title("Contour Plot of Leontief Utility Function")
ax2.set_xlabel("Good X")
ax2.set_ylabel("Good Y")
ax2.grid(False)

plt.tight_layout()
plt.show()
```

# Defining baskests and calculating utilities with both functions

```python
basket_1 = np.array([30, 60])
basket_2 = np.array([10, 20])
basket_3 = np.array([5, 10])
```

```python
from colored import Fore, Back, Style

U1_Cobb_Douglas = Cobb_Douglas(basket_1[0], basket_1[1])
U2_Cobb_Douglas = Cobb_Douglas(basket_2[0], basket_2[1])
U3_Cobb_Douglas = Cobb_Douglas(basket_3[0], basket_3[1])

print(f'{Fore.black}{Back.white} Utility of the first basket based on the Cobb Douglas utility function is {Fore.red}{U1_Cobb_Douglas}')
print(f'{Style.BOLD}{Fore.black}{Back.yellow} Utility of the second basket based on the Cobb Douglas utility function is {Fore.red}{U2_Cobb_Douglas} ')
print(f'{Style.underline_color(1)}{Fore.black}{Back.magenta} Utility of the third basket based on the Cobb Douglas utility function is {Fore.black}{U3_Cobb_Douglas} ')
```

```python
U1_leontief = Leontief(basket_1[0], basket_1[1])
U2_leontief = Leontief(basket_2[0], basket_2[1])
U3_leontief = Leontief(basket_3[0], basket_3[1])

print(f'{Fore.black}{Back.white} Utility of the first basket based on the Leontief utility function is {Fore.red}{U1_leontief}')
print(f'{Style.BOLD}{Fore.black}{Back.yellow} Utility of the second basket based on the Leontief utility function is {Fore.red}{U2_leontief} ')
print(f'{Style.underline_color(1)}{Fore.black}{Back.magenta} Utility of the third basket based on the Leontief utility function is {Fore.black}{U3_leontief} ')
```

### Cobb Douglas Satisfying Completeness:

```python
if U1_Cobb_Douglas > U2_Cobb_Douglas:
    print("Basket 1 is preferred over Basket 2")
elif U1_Cobb_Douglas < U2_Cobb_Douglas:
    print("Basket 2 is preferred over Basket 1")
else:
    print("Consumer is indifferent between Basket 1 and Basket 2")

if U1_Cobb_Douglas > U3_Cobb_Douglas:
    print("Basket 1 is preferred over Basket 3")
elif U1_Cobb_Douglas < U3_Cobb_Douglas:
    print("Basket 3 is preferred over Basket 1")
else:
    print("Consumer is indifferent between Basket 1 and Basket 3")

if U2_Cobb_Douglas > U3_Cobb_Douglas:
    print("Basket 2 is preferred over Basket 3")
elif U2_Cobb_Douglas < U3_Cobb_Douglas:
    print("Basket 3 is preferred over Basket 2")
else:
    print("Consumer is indifferent between Basket 2 and Basket 3")
```

## Satisfying Transitivity:

```python
def check_transitivity(U1, U2, U3):
    if U1 > U2 and U2 > U3:
        print("Transitivity is satisfied: U1 > U2 > U3")
    elif U1 < U2 and U2 < U3:
        print("Transitivity is satisfied: U1 < U2 < U3")
    else:
        print("Transitivity is not satisfied")
```

```python
result_of_cb = check_transitivity(U1_Cobb_Douglas, U2_Cobb_Douglas, U3_Cobb_Douglas)
print(f'Cobb_Douglas utility function satifies transitvity: {result_of_cb}')
```

```python
result_of_leon = check_transitivity(U1_leontief, U2_leontief, U3_leontief)
print(f'Leontief utility function satifies transitvity: {result_of_leon}')
```

## Satisfying Convexity:
**The mixture of two baskets is as good as one basket**

**I will check two baskets that have equivalent utilities**

### Cobb douglas

```python
basket_11 = np.array([8, 8])
basket_22 = np.array([4, 16])

U_basket_11 = Cobb_Douglas(basket_11[0], basket_11[1])
U_basket_22 = Cobb_Douglas(basket_22[0], basket_22[1])
print(f'utility of the first basket is {U_basket_11}')
print(f'utility of the second basket is {U_basket_22}')
```

```python
t = 0.5

mixed_x_from_basket1_and_basket2 = t * basket_11[0] + (1-t) * basket_22[0]
mixed_y_from_basket1_and_basket2 = t * basket_11[1] + (1-t) * basket_22[1]

U_mixed = Cobb_Douglas(mixed_x_from_basket1_and_basket2, mixed_y_from_basket1_and_basket2)

print(f'utility of the mixed basket is {U_mixed}')
print(f'convexity is satisfied (between mixture and first basket): {U_mixed >= U_basket_11}')
print(f'convexity is satisfied (between mixture and second basket): {U_mixed >= U_basket_22}')
```

### Leontief:

```python
eq_bundle1 = [30, 60]
eq_bundle2 = [30, 50]

U_eq_bundle1 = Leontief(eq_bundle1[0], eq_bundle1[1])
U_eq_bundle2 = Leontief(eq_bundle2[0], eq_bundle2[1])
print(f'utility of the first bundle is {U_eq_bundle1}')
print(f'utility of the second bundle is {U_eq_bundle2}')
```

```python
mixx = t * eq_bundle1[0] + (1-t) * eq_bundle2[0]
mixy = t * eq_bundle1[1] + (1-t) * eq_bundle2[1]

umix = Leontief(mixx, mixy)

print(f'utility of the mixed basket is {umix}')
print(f'convexity is satisfied (between mixture and first basket): {umix >= U_eq_bundle1}')
print(f'convexity is satisfied (between mixture and second basket): {umix >= U_eq_bundle2}')
```
