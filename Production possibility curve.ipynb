# Different Ways for Plotting the Production Possibility Curve in Python

## Part 1

### Constant trade-off between two goods

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
max_A = 100
max_B = 80

A = np.linspace(0, max_A, 100)
B = (max_B * (max_A - A)) / max_A

plt.figure(figsize=(10, 6))
plt.plot(A, B, label = "Production Possibility Curve", color = "red")
plt.fill_between(A, B, color = "lightblue", alpha = 0.5)
plt.title("Production Possibility Curve")
plt.xlabel("Good A Production")
plt.ylabel("Good B Production")
plt.ylim(0, max_A)
plt.xlim(0, max_B)
plt.grid(True)
plt.legend()
plt.show()
```

### Increasing trade-off between two goods

```python
max_C = 100
max_D = 80

C = np.linspace(0, max_C, 100)
D = max_B * (1 - (C / max_C) ** 2)

plt.figure(figsize=(10, 6))
plt.plot(C, D, label = "Production Possibility Curve", color = "red")
plt.fill_between(C, D, color = "lightblue", alpha = 0.5)
plt.title("Production Possibility Curve")
plt.xlabel("Good C Production")
plt.ylabel("Good D Production")
plt.ylim(0, max_C)
plt.xlim(0, max_D)
plt.grid(True)
plt.axhline(0, color = "black", linewidth = 0.5, ls = "--")
plt.axvline(0, color = 'black', linewidth = 0.5, ls = '--')
plt.legend()
plt.show()
```

## How can we produce a range of numeric data type?

```python
import numpy as np
```

np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)

Parameters

start: The starting value of the sequence.

stop: The end value of the sequence.

num: (Optional) The number of samples to generate. Default is 50.

endpoint: (Optional) If True (default), stop is the last sample. If False, it is not included.

retstep: (Optional) If True, returns the samples and the step size between them.

dtype: (Optional) The desired data type of the output array


```python
# Generate 5 evenly spaced numbers between 0 and 1
s = np.linspace(0, 1, num=5)
print(s)
```

```python
ss = np.linspace(0, 1, num=5, endpoint=False)
print(ss)
```

# Plotting the PPC

We can have a maximum of two goods, and producing each of them comes at a cost. We must give up a certain amount of one good to produce another. Therefore, they are functions of each other. For simplicity, we will assume that only one good is a function of the other.

# Constant trade-off

```python
max_A = 100
max_B = 80

A = np.linspace(0, max_A, 100)
B = (max_B * (max_A - A)) / max_A
```

```python
plt.figure(figsize=(10, 6))
plt.plot(A, B, label = "Production Possibility Curve", color = "red")
plt.fill_between(A, B, color = "lightblue", alpha = 0.5)
plt.title("Production Possibility Curve")
plt.xlabel("Good A Production")
plt.ylabel("Good B Production")
plt.ylim(0, max_A)
plt.xlim(0, max_B)
plt.grid(True)
plt.legend()
plt.show()
```

# increasing trade-off

```python
max_C = 100
max_D = 80

C = np.linspace(0, max_C, 100)
D = max_B * (1 - (C / max_C) ** 2)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(C, D, label = "Production Possibility Curve", color = "red")
plt.fill_between(C, D, color = "lightblue", alpha = 0.5)
plt.title("Production Possibility Curve")
plt.xlabel("Good C Production")
plt.ylabel("Good D Production")
plt.ylim(0, max_C)
plt.xlim(0, max_D)
plt.grid(True)
plt.axhline(0, color = "black", linewidth = 0.5, ls = "--")
plt.axvline(0, color = 'black', linewidth = 0.5, ls = '--')
plt.legend()
plt.show()
```
