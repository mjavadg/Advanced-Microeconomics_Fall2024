# Axioms of rationality part 3

$$ 
\text{Consumption Plan or Consumption Bundle: } \mathbf{x} = (x_1, x_2, \dots, x_n)
$$

$$
\text{A consumption bundle } \mathbf{x} \in X \text{ is thus represented by a point } \mathbf{x} \in R^n_{+}.
$$

$$ \text{ So } X = R^n_{+} $$

## Properties of The Consumption set, X

The minimal requirements on the consumption are

1. $$ X \subseteq R^n_{+}$$

2. $$ X \text{ is closed} $$

3. $$ X \text{ is convex} $$

4. $$ \mathbf{0} \in X $$

Feasible set $\mathit{B}$ is that subset of the consumption set $\mathit{X}$ that remains after we have accounted for any constraints on the consumer's access to commodities due to the practical, institutional, or economic realities of the world. To put it simply, assume $\mathit{B} \subset \mathit{X}$.

A **preference relation** typically specifies the limits, if any, on the consumer’s ability to perceive in situations involving choice the form of consistency or inconsistency in the consumer’s choices, and information about the consumer’s tastes for the different objects of choice.

Finally, the model is ‘closed’ by specifying some **behavioural assumption**. This expresses the guiding principle the consumer uses to make final choices and so identifies the ultimate objectives in choice. *It is supposed that the consumer seeks to identify and select an available alternative that is most preferred in the light of his personal tastes.*

**axioms of consumer** choice formalise the view that the consumer can choose and that choices are consistent in a particular way.

Formally, we represent the consumer’s preferences by a binary relation, $\succsim$, defined on the consumption set, $X$. If $x^1 \succsim x^2$, we say that ‘$x^1$ is at least as good as $x^2$’ for this consumer.

$$
\text{Axiom 1: Completeness. For all } x^1 \text{ and } x^2 \text{ in } X, \text{ either } x^1 \succsim x^2 \text{ or } x^2 \succsim x^1.
$$


$$
\text{Axiom 2: Transitivity. For any three elements } x^1, x^2 \text{ and } x^3 \text{ in } X, \text{if } x^1 \succsim x^2 \text{ and } x^2 \succsim x^3, \text{ then } x^1 \succsim x^3
$$

### Definition 1.1 Preference Relation

*The binary relation $\succsim$ on the consumption set X is called a preference relation if it satisfies Axioms 1 and 2.*


### Definition 1.2 Strict Preference Relation

*The binary relation $\succ$ on the consumption set X is defined as follows:*

$$ 
\mathbf{x}^1 \succ \mathbf{x}^2 \mathit{\text{ if and only if }} x^1 \succsim x^2 \text{ and } x^2 \nsucceq x^1.
$$

*The relation $\succ$ is called the strict preference relation induced by $\succsim$ , or simply the strict preference relation when $\succsim$ is clear. The phrase $\mathbf{x}^1$ $\succ$ $\mathbf{x}^2$ is read, ‘$\mathbf{x}^1$ is strictly preferred to $\mathbf{x}^2$’.*

### Definition 1.3 Indifference Relation

*The binary relation $\sim$ on the consumption set X is defined as follows:*

$$
\mathbf{x}^1 \sim \mathbf{x}^2 \mathit{\text{ if and only if }} x^1 \succsim x^2 \text{ and } x^2 \succsim x^1.
$$

*The relation $\sim$ is called the indifference relation induced by $\succsim$ , or simply the indifference
relation when $\succsim$ is clear. The phrase $\mathbf{x}^1$ $\sim$ $\mathbf{x}^2$ is read, ‘$\mathbf{x}^1$ is indifferent to $\mathbf{x}^2$’*


$$
\text{Axiom 3: Continuity. For all } \mathbf{x} \in R^n_{+}, \text{the 'at least as good as' set, } \succsim \mathbf{(x)}, \text{and the 'no better than' set}, \precsim \mathbf{(x)} \text{ are closed in } R^n_{+}.
$$

## Definition:

$\{x \in X \mid x \succeq y\}$ called the 'at least as good as' set.

$\{x \in X \mid x \preceq y\}$ called the 'no better than' set.

$\{x \in X \mid x \prec y\}$ called the 'worse than' set.

$\{x \in X \mid x \succ y\}$ called the 'preferred to' set.

$\{x \in X \mid x \sim y\}$ called the 'indifference' set.


$$

$$

$$ 
\text{Axiom 4: Strict Monotonicity. For all } \mathbf{x}^0, \mathbf{x}^1 \in R^n_{+}, \text{ if } \mathbf{x}^0 \geq \mathbf{x}^1 \text{ then } \mathbf{x}^0 \succsim \mathbf{x}^1, \text{ while if } \mathbf{x}^0 \gg \mathbf{x}^1, \text{ then } \mathbf{x}^0 \succ \mathbf{x}^1. 
$$

$$
\text{Axiom 5': Convexity. if } \mathbf{x}^0 \succsim \mathbf{x}^1, \text{ then } t\mathbf{x}^1 + (1 - t)\mathbf{x}^0 \succsim \mathbf{x}^0 \text{ for all } t \in [0, 1]. 
$$

$$
\text{Axiom 5: Strict Convexity. if } \mathbf{x}^1 \neq \mathbf{x}^0 \text{ and } \mathbf{x}^1 \succsim \mathbf{x}^0, \text{ then } t\mathbf{x}^1 + (1 - t)\mathbf{x}^0 \succ \mathbf{x}^0 \text{ for all } t \in (0, 1). 
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

# Define the set of choices to form a convex indifference curve
set_of_choices = [(0, 10), (1, 8), (2, 6.5), (3, 5.5), (4, 5), (5, 4.8)]

# Extract x and y coordinates
x, y = zip(*set_of_choices)

# Sort the coordinates based on x values
sorted_pairs = sorted(zip(x, y))
x_sorted, y_sorted = zip(*sorted_pairs)

# Generate a smooth curve
x_smooth = np.linspace(min(x_sorted), max(x_sorted), 300)  # More points for a smoother curve
spl = make_interp_spline(x_sorted, y_sorted, k=3)  # B-spline with degree 3
y_smooth = spl(x_smooth)

# Plot the points and the smooth curve
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='#549A40')
plt.plot(x_smooth, y_smooth, color='blue', linestyle='-')

plt.plot([1, 3], [8, 5.5], linestyle='--', color='red')
plt.scatter(1.72, 7.1, color='#549A40')
plt.annotate(r'$\mathbf{x}^t$', (1.72, 7.1), textcoords="offset points", xytext=(0,8), ha='center', fontsize=12, color='black')

plt.annotate(r'$\mathbf{x}^1$', (3, 5.5), textcoords="offset points", xytext=(0,10), ha='center', fontsize=12, color='black')
plt.annotate(r'$\mathbf{x}^2$', (1, 8), textcoords="offset points", xytext=(0,10), ha='center', fontsize=12, color='black')

plt.plot([4, 2.5], [5, 5.5], linestyle='-', color='black')
plt.plot([0.5, 1.2], [8.91, 7.5], linestyle='-', color='black')


# Set the title and labels
plt.title("Indifference Curve")
plt.xlabel('Number of first goods in bundles')
plt.ylabel('Number of second goods in bundles')

# Grid and legend adjustments
plt.grid(True)
plt.legend(loc='upper left')  # Position the legend at the top left

# Show the plot
plt.show()
```
