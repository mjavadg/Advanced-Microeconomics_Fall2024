## Transitivity

- Transitivity. If an individual reports that A is preferred to B and B is preferred to C, then he or she must also report that A is preferred to C.

```python
#approach = summation

sets1 = [(3, 2), (3, 1), (2, 1)]

utility_a = sum(sets1[0])
utility_b = sum(sets1[1])
utility_c = sum(sets1[2])


print(utility_a, utility_b, utility_c)

if utility_a > utility_b and utility_b > utility_c and utility_a > utility_c:
    print("transitivity is satisfied")
```

## Continuity

- Continuity. If an individual reports A is preferred to B, then situations suitably close to A must also be preferred to B.

```python
#approach = summation

bundle_A = (3, 4)
bundle_B = (2, 5)
a_close_bundle = (3.1, 4.1) #a bundle close to A

utility_A = sum(bundle_A)
utility_B = sum(bundle_B)
utility_close_to_A = sum(a_close_bundle)

is_preferred_A_B = utility_A > utility_B
is_close_to_A = True #assume this is true
is_preferred_close_A_B = utility_close_to_A > utility_B

print("Bundle A is preferred over bundle B: ", is_preferred_A_B)
print("Close bundle to A is preferred over bundle B:", is_preferred_close_A_B)
```
