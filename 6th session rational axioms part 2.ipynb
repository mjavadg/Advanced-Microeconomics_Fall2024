## Transitivity

- Transitivity. If an individual reports that ‘‘A is preferred to B’’ and ‘‘B is preferred to C,’’ then he or she must also report that ‘‘A is preferred to C.’’

```python
#approach = summation

sets1 = [(3, 2), (3, 1), (2, 1)]

utility_a = sum(sets1[0])
utility_b = sum(sets1[1])
utility_c = sum(sets1[2])


print(utility_a, utility_b, utility_c)

if utility_a > utility_b and utility_b > utility_c and utility_a > utility_c:
    print("transitivity is satisfied")
```

# Plotting indifference curves based on different axioms.

```python
# Define the set of choices
set_of_choices = [(3, 2), (2, 1), (3, 1)]

# Extract x and y coordinates
x, y = zip(*set_of_choices)
print(x, y)

# Plot the points and the indifference curve
plt.plot(x, y, marker = "o", color = '#549A40', 
         label = "Indifference curve based on\ncompleteness and transitivity axioms", linestyle = '--')

# Set the title and labels
plt.title("Locations of Bundles")
plt.xlabel('Number of first goods in bundles')
plt.ylabel('Number of second goods in bundles')

# Grid and legend adjustments
plt.grid(True)
plt.legend(loc='upper left')  # Position the legend at the top left

# Show the plot
plt.show()
```

### In the graph above, the dashed line represents the possibility of non-continuous bundles, indicating that the bundles do not form a smooth, continuous curve but rather discrete points that may not be directly connected.

### The graph below shows a better representation. In this graph, some bundles are continuous (generate the same level of utility) and some are not.

```python
# Define the set of choices
set_of_choices = [(0, 10), (1, 9), (1.7, 8.6), (1.4, 8.3), (2, 8), (3, 7), (3.1, 6.9)]

# Extract x and y coordinates
x, y = zip(*set_of_choices)
print(x, y)

# Create a new figure
plt.figure(figsize=(10, 6))

# Plot discrete points
plt.scatter(x, y, color='#549A40', label="Bundles")

# Plot dashed lines for non-continuous connections
plt.plot([1, 1.7], [9, 8.6], linestyle='--', color='blue')
plt.plot([1, 1.4], [9, 8.3], linestyle='--', color='blue')
plt.plot([1.7, 2], [8.6, 8], linestyle='--', color='blue')
plt.plot([1.4, 2], [8.3, 8], linestyle='--', color='blue')

# Plot continuous lines for continuous connections
plt.plot([0, 1], [10, 9], linestyle = '-', color = 'blue')
plt.plot([2, 3], [8, 7], linestyle='-', color='blue')
plt.plot([3, 3.1], [7, 6.9], linestyle='-', color='blue')

# Set the title and labels
plt.title("Indifference curve based on\ncompleteness and transitivity axioms")
plt.xlabel('Number of first goods in bundles')
plt.ylabel('Number of second goods in bundles')

# Grid and legend adjustments
plt.grid(True)
plt.legend(loc='upper left')  # Position the legend at the top left

# Show the plot
plt.show()
```

## Continuity

- Continuity. If an individual reports ‘‘A is preferred to B,’’ then situations suitably ‘‘close to’’ A must also be preferred to B.

```python
#approach = summation

bundle_A = (3, 4)
bundle_B = (2, 5)
a_close_bundle = (3.1, 4.1) #a bundle close to A

utility_A = sum(bundle_A)
utility_B = sum(bundle_B)
utility_close_to_A = sum(a_close_bundle)

is_preferred_A_B = utility_A > utility_B
is_close_to_A = True #assume this is true
is_preferred_close_A_B = utility_close_to_A > utility_B

print("Bundle A is preferred over bundle B: ", is_preferred_A_B)
print("Close bundle to A is preferred over bundle B:", is_preferred_close_A_B)
```

## Plotting a difference curve based on completeness, transitivity, and continuity

```python
# Define the set of choices
set_of_choices = [(0, 10), (1, 9), (1.7, 8.6), (1.4, 8.3), (2, 8), (3, 7), (3.1, 6.9)]

# Extract x and y coordinates
x, y = zip(*set_of_choices)
print(x, y)

# Create a new figure
plt.figure(figsize=(10, 6))

# Plot discrete points
plt.scatter(x, y, color='#549A40', label="Bundles")

# Plot dashed lines for non-continuous connections
plt.plot([1, 1.7], [9, 8.6], linestyle='-', color='blue')
plt.plot([1, 1.4], [9, 8.3], linestyle='-', color='blue')
plt.plot([1.7, 2], [8.6, 8], linestyle='-', color='blue')
plt.plot([1.4, 2], [8.3, 8], linestyle='-', color='blue')

# Plot continuous lines for continuous connections
plt.plot([0, 1], [10, 9], linestyle = '-', color = 'blue')
plt.plot([2, 3], [8, 7], linestyle='-', color='blue')
plt.plot([3, 3.1], [7, 6.9], linestyle='-', color='blue')

# Set the title and labels
plt.title("Indifference curve based on\ncompleteness, transitivity, and continuity axioms")
plt.xlabel('Number of first goods in bundles')
plt.ylabel('Number of second goods in bundles')

# Grid and legend adjustments
plt.grid(True)
plt.legend(loc='upper left')  # Position the legend at the top left

# Show the plot
plt.show()
```

# Pay attention please:

*Up to this point, we have successfully captured the ideas of completeness, transitivity, and continuity in our analysis of consumer preferences. The indifference curves we have plotted are represented as straight lines, reflecting a strict interpretation of these axioms.*

•Completeness ensures that consumers can compare any two bundles of goods and express a preference or indifference between them.

•Transitivity guarantees that if a consumer prefers bundle A to bundle B and bundle B to bundle C, then they must prefer bundle A to bundle C.

•Continuity implies that small changes in the consumption of goods will not lead to abrupt changes in utility.

*The straight-line representation of indifference curves indicates that we have not yet imposed the axiom of convexity. This means that while our current model satisfies the aforementioned axioms, it does not account for the diminishing marginal rates of substitution typically observed in real-world preferences.In upcoming sections, we will delve deeper into additional axioms, like **monotonocity and convexity**, and explore how they can alter the shape of indifference curves to better reflect rational consumer behavior. This exploration will enhance our understanding of how preferences can be modeled more realistically.*
